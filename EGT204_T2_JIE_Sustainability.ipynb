```python
# Drop the year 2023
gdp_singapore = gdp_singapore.drop(columns=[2023])

# Drop the years between 1960 and 1970
years_to_drop = list(range(1960, 1970))
gdp_singapore = gdp_singapore.drop(columns=years_to_drop)
gdp_singapore
```

## Prepare the datasets to find the correlation. Can use linear regression

```python
#list of all the gdp values from year 1970 to 2022 in gdp_singapore
gdp_values = gdp_singapore.loc[:, 1970:2022].values.tolist()[0]
gdp_values
```

```python
#list of all the co2 values from year 1970 to 2022 in co2_yearly_singapore
co2_values = co2_yearly_singapore.loc[:, '1970':'2022'].values.tolist()[0]
co2_values
```

```python
#Create a scatterplot and plot the best fit line
from scipy.stats import linregress

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(gdp_values, co2_values)

# Create the best fit line
line = [slope * xi + intercept for xi in gdp_values]

# Plot the scatter plot and best fit line
plt.scatter(gdp_values, co2_values)
plt.plot(gdp_values, line, color='red')
plt.xlabel('GDP')
plt.ylabel('CO2 Emissions')
plt.title('Relationship between GDP and CO2 Emissions in Singapore')
plt.show()
```

```python
# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(gdp_values, co2_values)[0, 1]
print("Correlation Coefficient:", correlation_coefficient)
```

```python
#Grouping the countries into the 7 different continents
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2

# Function to get continent code
def get_continent(country_name):
  try:
    country_alpha2 = country_name_to_country_alpha2(country_name)
    continent_code = country_alpha2_to_continent_code(country_alpha2)
    return continent_code
  except:
    return 'Unknown'

# Apply the function to the 'Country' column
co2_per_capita_cleaned['Continent'] = co2_per_capita_cleaned['Country'].apply(get_continent)
co2_per_capita_cleaned.head()
```

```python
#create a column called total where it is the sum of all the sectors except for country
sg_sectors_2022_pivot['Total'] = sg_sectors_2022_pivot.drop('Country', axis=1).sum(axis=1)
sg_sectors_2022_pivot
```

```python
# Calculate the percentage contribution of each sector
for sector in sg_sectors_2022_pivot.columns[1:-1]:  # Exclude 'Country' and 'Total' columns
  sg_sectors_2022_pivot[sector] = (sg_sectors_2022_pivot[sector] / sg_sectors_2022_pivot['Total']) * 100
sg_sectors_2022_pivot
```

```python
# transform the data where the sectors is the columns and a year column

# Melt the DataFrame to long format
sg_sectors_long = sg_sectors_10years.melt(
    id_vars=['Country', 'Sector', 'Substance'],
    var_name='Year',
    value_name='Emissions'
)

# Pivot the DataFrame to make sectors as columns
sg_sectors_pivot = sg_sectors_long.pivot_table(
    index=['Country', 'Year'],
    columns='Sector',
    values='Emissions',
    aggfunc='sum'  # Sum emissions if multiple substances per sector
).reset_index()

# Display the transformed DataFrame
display(sg_sectors_pivot)
```

```python
# create a column called top 3 sectors where its the sum of Industrial Combustion processes and power industry

# Calculate the sum of emissions for the top 3 sectors
sg_sectors_pivot['Top 3 Sectors'] = sg_sectors_pivot['Industrial Combustion'] + sg_sectors_pivot['Processes'] + sg_sectors_pivot['Power Industry']

# Display the DataFrame with the new column
display(sg_sectors_pivot)
```

```python
# Remove the last row
sg_sectors_pivot = sg_sectors_pivot[:-1]

# Display the updated DataFrame
display(sg_sectors_pivot)
```
